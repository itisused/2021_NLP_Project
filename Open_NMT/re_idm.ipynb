```python
import re
```

```python
src_test = "/content/drive/MyDrive/08_TeamProject/data/4.AIHub_KISS/src-test.txt"
src_train = "/content/drive/MyDrive/08_TeamProject/data/4.AIHub_KISS/src-train.txt"
src_val = "/content/drive/MyDrive/08_TeamProject/data/4.AIHub_KISS/src-val.txt"
```

```python
file_list = [src_test, src_train, src_val]
```

```python
for var in file_list:
    new_corpus = []
    with open(var, "r") as file:
        for line in file:
            s = re.sub("< idm >", "<idm>", line)
            new_corpus.append(s)
    res = re.sub("src-", "src-re-", var)
    with open(res, "w") as file:
        file.writelines(new_corpus)
```
