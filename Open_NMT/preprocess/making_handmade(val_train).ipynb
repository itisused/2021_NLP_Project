```python
import pandas as pd
from tqdm import tqdm
```

```python
idx_list = []
# f01 = open("/home/lab19/OpenNMT_latest/data/src-re-train.txt", "r")
f01 = open("/home/lab19/OpenNMT_latest/data/src-re-val.txt", "r")
f_list = f01.readlines()
for idx, line in tqdm(enumerate(f_list)):
    if "<idm>" in line:
        idx_list.append(idx)
```

```python
# f02 = open("/home/lab19/kiss_test/data/2.AIHub_KISS/src-train.txt", "r")
f02 = open("/home/lab19/kiss_test/data/2.AIHub_KISS/src-val.txt", "r")
s = f02.readlines()
for idx in idx_list:
    str01 = s[idx]
    str02 = "<idm> "
    s[idx] = str02 + str01
    print(s[idx])
```

```python
with open("/home/lab19/kiss_test/data/5.Handmade/handmade_src_val.txt", "w") as f:
    f.writelines(s)
```
