```python
import pandas as pd
```

```python

# 한국어 train 데이터를 열어주고, 개행문자를 지워줌, length를 뽑아서 한국어와 영어 파일의 길이가 같은지 살펴보기
f1 = open("G:/GoogleDrive_Coding/python_jobs/nlp_teamplay/kiss/data/final_corpus/src-train.txt", "r", encoding="utf-8")
lines = f1.readlines()
ko = []

for line in lines:
    line = line.replace("\n","")
    line = line.strip()
    ko.append(line)

print(len(ko))
```

```python
# 영어 train 데이터를 열어주고, 개행문자를 지워줌, length를 뽑아서 한국어와 영어 파일의 길이가 같은지 살펴보기
f2 = open("G:/GoogleDrive_Coding/python_jobs/nlp_teamplay/kiss/data/final_corpus/tgt-train.txt", "r", encoding="utf-8")
lines = f2.readlines()
en = []

for line in lines:
    line = line.replace("\n","")
    line = line.strip()
    en.append(line)

print(len(en))
```

```python
# Classifier가 관용구로 분류한 KCC940 한국어 문장들을 불러와 개행문자를 없애고 <idm> 토큰을 붙임
f3 = open("G:/GoogleDrive_Coding/python_jobs/nlp_teamplay/data/kcc940_idiom_pos.txt", "r", encoding="utf-8")
lines = f3.readlines()
ko_news = []

for line in lines:
    line = line.replace("\n","")
    line = "<idm> "+line
    line = line.strip()
    ko_news.append(line)

print(len(ko_news))
```

```python
# 관용구로 분류된 KCC940 한국어 문장들을 80000번 학습시킨 모델에 넣어 prediction한 결과물(영어)을 불러오고, 개행문자를 없앰
f4 = open("G:/GoogleDrive_Coding/python_jobs/nlp_teamplay/data/kcc_pred_80000.txt", "r", encoding="utf-8")
lines = f4.readlines()
en_news = []

for line in lines:
    line = line.replace("\n","")
    line = line.strip()
    en_news.append(line)

print(len(en_news))
```

```python
# 기존 한국어 train 데이터를 데이터프레임으로 바꿔줌
ko_df = pd.DataFrame(ko)
ko_df.columns=["ko"]
ko_df
```

```python
# 기존 영어 train 데이터를 데이터프레임으로 바꿔줌
en_df = pd.DataFrame(en)
en_df.columns=["en"]
en_df
```

```python
# 앞에 idm 토큰을 붙인 KCC940 한국어 문장을 데이터프레임으로 바꿔줌
ko_df_news = pd.DataFrame(ko_news)
ko_df_news.columns=["ko"]
ko_df_news
```

```python
# KCC940 관용구를 prediction한 영어 문장들을 데이터프레임으로 바꿔줌
en_df_news = pd.DataFrame(en_news)
en_df_news.columns=["en"]
en_df_news
```

```python
# 기존 한국어 train 데이터와 <idm> 태깅한 KCC940 관용구 데이터를 concat
ko_result = pd.concat([ko_df, ko_df_news], ignore_index=True)
ko_result
```

```python
# 기존 영어 train 데이터와 KCC940 prediction한 영어 문장들 concat
en_result = pd.concat([en_df, en_df_news], ignore_index=True)
en_result
```

```python
# 한국어 데이터프레임과 영어 데이터프레임을 concat
new_train = pd.concat([ko_result, en_result], axis=1)
new_train
```

```python
# 셔플링으로 순서를 섞어줌
final_df = new_train.sample(frac=1).reset_index(drop=True)
final_df
```

```python
# 전체 데이터프레임을 csv 형식으로 저장
final_df.to_csv('new_train_corpus.csv', index=False)
```

```python
# 한글만 src-new-train.txt로 저장
with open('src-new-train.txt', 'w', encoding="utf-8") as f_ko: 
    for line in final_df['ko']:
        f_ko.write(line + '\n')
        # print(line)
```
