```python
general_corpus = pd.read_csv('./data/final_corpus/general_corpus.csv')
general_corpus
```

```python
idiom_corpus = pd.read_csv('./data/final_corpus/idiom_corpus.csv')
idiom_corpus = idiom_corpus.sample(frac=1)
idiom_corpus
```

# Train corpus 생성

```python
train_df = pd.concat([general_corpus[:1564450], idiom_corpus[:5855]])
train_df = train_df.sample(frac=1)
```

```python
train_df
```

```python
train_df.to_csv('./data/final_corpus/train_corpus.csv', index=False)
```

```python
with open('./data/final_corpus/src-train.txt', 'w') as f: 
    for k in train_df['ko']:
        f.write(k + '\n')
        # print(k)
```

```python
with open('./data/final_corpus/tgt-train.txt', 'w') as f: 
    for e in train_df['en']:
        f.write(e + '\n')
```

# Validaion corpus 생성

```python
val_df = pd.concat([general_corpus[1564450:1564450+4500], idiom_corpus[5855:5855+500]])
val_df = val_df.sample(frac=1)
val_df
```

```python
val_df.to_csv('./data/final_corpus/val_corpus.csv', index=False)
```

```python
with open('./data/final_corpus/src-val.txt', 'w') as f: 
    for k in val_df['ko']:
        f.write(k + '\n')
```

```python
with open('./data/final_corpus/tgt-val.txt', 'w') as f: 
    for e in val_df['en']:
        f.write(e + '\n')
```

# Test Corpus 생성

```python
test_df = pd.concat([general_corpus[1564450+4500:], idiom_corpus[5855+500:]])
test_df = test_df.sample(frac=1)
test_df
```

```python
test_df.to_csv('./data/final_corpus/test_corpus.csv', index=False)
```

```python
with open('./data/final_corpus/src-test.txt', 'w') as f: 
    for k in test_df['ko']:
        f.write(k + '\n')
```

```python
with open('./data/final_corpus/tgt-test.txt', 'w') as f: 
    for e in test_df['en']:
        f.write(e + '\n')
```
