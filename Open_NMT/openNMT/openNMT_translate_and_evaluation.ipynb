# Encoding SPM

```python
enc_model_prefix = "/home/lab03/src/openNMT_final/data/subword1/subword_kor"
enc_sp = spm.SentencePieceProcessor()
enc_sp.Load(enc_model_prefix + '.model')
```

```python
with open('/home/lab03/src/openNMT_final/data/kcc940_idiom_pos.txt', 'r') as f:
    src_test = f.readlines()
```

```python
spm_enc = []
for t in src_test[:10]:
    enc = enc_sp.encode_as_pieces(t)
    print(" ".join(enc))
    # print(enc)
    # spm_enc.append(" ".join(enc))
```

```python
with open('/home/lab03/src/openNMT_final/data/kcc_test_spm_500.txt', 'w') as f:
    for d in spm_enc:
        f.write(d + '\n')
```

# Decoding Prediction

```python
dec_model_prefix = "/home/lab03/src/openNMT_final/data/subword1/subword_eng"
dec_sp = spm.SentencePieceProcessor()
dec_sp.Load(dec_model_prefix + '.model')
```

```python
# onmt_translate output file open
with open('/home/lab03/src/openNMT_final/data/predict/spred_100000_spm.txt', 'r') as f:
    data = f.readlines()
```

```python
spm_dec = []
for d in data:
    dec = dec_sp.decode_pieces(d.split())
    spm_dec.append(dec)
```

```python
with open('/home/lab03/src/openNMT_final/data/predict/dec_pred_80000.txt', 'w') as f:
    for d in spm_dec:
        f.write(d + '\n')
```
